```python
# Gerekli kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Rastgele bir veri seti oluştur
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Veri setini görselleştirme
plt.scatter(X[:, 0], X[:, 1], s=50, c='gray', marker='o')
plt.title("Rastgele Oluşturulmuş Veri Seti")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.show()

# 2. Dirsek yöntemi ile optimal k değeri bulma
inertia = []  # Inertia değerlerini saklayacağımız liste

for k in range(1, 11):  # Küme sayısını 1'den 10'a kadar dene
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Inertia değerlerini görselleştirme
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Dirsek Yöntemi ile Optimal Küme Sayısı")
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Inertia")
plt.show()

# 3. Optimal küme sayısını belirle ve K-Means'i uygula
optimal_k = 4  # Dirsek yöntemine göre belirlenen optimal k değeri
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)

# Küme etiketlerini ve küme merkezlerini al
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# 4. Kümelenmiş veriyi görselleştirme
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X', label="Merkezler")
plt.title(f"K-Means Sonucu (k={optimal_k})")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# 1. Rastgele bir veri seti oluştur
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Veri setini görselleştirme
plt.scatter(X[:, 0], X[:, 1], s=50, c='gray', marker='o')
plt.title("Rastgele Oluşturulmuş Veri Seti")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.show()

# 2. Optimal küme sayısını BIC ile belirleme
bic_scores = []
aic_scores = []
components = range(1, 11)

for k in components:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(X)
    bic_scores.append(gmm.bic(X))
    aic_scores.append(gmm.aic(X))

# 3. BIC ve AIC skorlarını görselleştirme
plt.plot(components, bic_scores, label='BIC', marker='o')
plt.plot(components, aic_scores, label='AIC', marker='o')
plt.xlabel("Küme Sayısı (n_components)")
plt.ylabel("Skor")
plt.title("BIC ve AIC ile Optimal Küme Sayısı")
plt.legend()
plt.show()

# 4. GMM ile kümeleme (optimal küme sayısını seçerek)
optimal_k = 4  # Grafikten çıkarılan optimal değer
gmm = GaussianMixture(n_components=optimal_k, random_state=42)
gmm.fit(X)

# Küme tahminleri ve parametreler
labels = gmm.predict(X)
means = gmm.means_
covariances = gmm.covariances_

# 5. Sonuçları görselleştirme
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(means[:, 0], means[:, 1], s=300, c='red', marker='X', label='Merkezler')
plt.title(f"GMM Sonucu (k={optimal_k})")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.legend()
plt.show()
```

```python
# Gerekli kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# 1. Karmaşık şekilli bir veri seti oluştur
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

# Veri setini görselleştirme
plt.scatter(X[:, 0], X[:, 1], s=50, c='gray', marker='o')
plt.title("Karmaşık Şekilli Veri Seti (Make Moons)")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.show()

# 2. DBSCAN ile kümeleme
dbscan = DBSCAN(eps=0.2, min_samples=5)  # eps ve min_samples parametrelerini ayarla
dbscan.fit(X)

# Küme etiketlerini al
labels = dbscan.labels_

# 3. Kümelenmiş veriyi görselleştirme
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title("DBSCAN ile Kümeleme")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.show()

# Gürültü noktalarını ayrı görselleştirme
noise_points = X[labels == -1]
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, label='Kümeler')
plt.scatter(noise_points[:, 0], noise_points[:, 1], c='red', label='Gürültü')
plt.title("Gürültü Noktaları ve Kümeler")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
nbrs = NearestNeighbors(n_neighbors=5).fit(X)
distances, indices = nbrs.kneighbors(X)

distances = np.sort(distances[:, 4])  # En uzak 5. komşu mesafesi
plt.plot(distances)
plt.title("K-En Yakın Komşu Mesafesi Grafiği")
plt.xlabel("Veri Noktaları")
plt.ylabel("Mesafe")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# 1. Gürültülü bir veri seti oluştur
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Gürültü noktaları ekle
np.random.seed(42)
noise = np.random.uniform(low=-10, high=10, size=(20, 2))  # Gürültü noktaları
X = np.vstack([X, noise])  # Gürültüyü veri setine ekle

# Veri setini görselleştirme
plt.scatter(X[:, 0], X[:, 1], s=50, c='gray', marker='o')
plt.title("Gürültülü Veri Seti")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.show()

# 2. DBSCAN ile kümeleme
dbscan = DBSCAN(eps=1.2, min_samples=5)  # eps ve min_samples parametrelerini ayarla
dbscan.fit(X)

# Küme etiketlerini al
labels = dbscan.labels_

# 3. Gürültü noktalarını işaretle (label -1)
core_samples_mask = labels != -1  # Çekirdek ve kenar noktalar
noise_mask = labels == -1  # Gürültü noktaları

# Sonuçları görselleştirme
plt.scatter(X[core_samples_mask, 0], X[core_samples_mask, 1], c=labels[core_samples_mask], cmap='viridis', s=50, label='Kümeler')
plt.scatter(X[noise_mask, 0], X[noise_mask, 1], c='red', s=50, label='Gürültü')
plt.title("DBSCAN ile Kümeleme ve Gürültü Tespiti")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.legend()
plt.show()
```
